```python
"""Oppgave 5"""

print("Dette programmet regner ut areal og ytre omkrets av følgende figur:")

import sys
sys.path.append("PY1010")           #Utvider søkestien

import matplotlib.pyplot as plt     
import matplotlib.image as mpimg    #importerer to moduler for bildevisning

bilde = mpimg.imread("FIGUR.png")  #angir hvilken fil image-modulen skal lese
plt.imshow(bilde)  #klargjør aktuelt bilde for visning
plt.axis("off")    #skrur av aksene
plt.show()         #viser bildet på skjerm

import math        #inn med nødvendig modul for følgende beregninger

a = 4              #angir variabler for følgende funksjon
b = 8

def figur_areal_ytreomkrets(a,b):
    #Blokkommentar som forklarer funksjonen:
    """Denne funksjonen regner ut areal og ytre omkrets av figuren over.
    """  
    c = math.sqrt (a**2 + b**2)                  #definererer en variabel som trengs i utregningen
    areal = math.pi * (a/2)**2 / 2 + (b*a)/2     #regner ut arealet
    ytre_omkrets = math.pi * (a / 2) + b + c     #regner ut ytre omkrets
    return areal, ytre_omkrets                   #gir areal og ytre_omkrets som output

res1 = figur_areal_ytreomkrets(a,b)[0]           #deklarerer en variabel for å vise resultat areal
res2 = figur_areal_ytreomkrets(a,b)[1]           #deklarerer en variabel for å vise resultat ytre_omkrets

print("La oss si at a =", a,"cm og b =", b,"cm. Da er figurens areal ", round(res1, 1),"cm2, " 
      "og den ytre omkretsen er", round(res2, 1), "cm.") 
                                                 #skriver en setning til skjerm som viser grunnlag for utregning og resultatet.
                                                 #bruker round med en desimal for ryddig resultat

```
